```python
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def func(delta, alpha, beta):
    return alpha / (beta + delta)
```

```python
delta = np.linspace(0.0, 10.0, 100)


alpha_betas = [
    (1.0, 2.0, '-'),
    (1.0, 1.0, '-'),
    (1.0, 0.5, '-'),
    (2.0, 4.0, '--'),
    (2.0, 2.0, '--'),
    (2.0, 1.0, '--'),
]

plt.figure(figsize=(10, 6))
for a, b, ls in alpha_betas:
    plt.plot(delta, func(delta, alpha=a, beta=b), ls=ls, lw=2,
             label=rf"$\alpha={a:.1f} ~~~ \beta={b:.1f}$")
plt.legend();
```
